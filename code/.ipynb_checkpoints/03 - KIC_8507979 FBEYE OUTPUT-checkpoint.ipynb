```python
import pymc3 as pm
import pymc3_ext as pmx
import aesara_theano_fallback.tensor as tt
from celerite2.theano import terms, GaussianProcess
```

```python
# for each Kepler Qtr:
# mask flares
# fit GP
# flatten LC
# compute new EDs
```

```python
fmask = np.ones_like(lc['t'].values, dtype=int)
for i in range(len(ff)):
    fmask[int(ff['Start_INDX'].values[i]):int(ff['Stop_INDX'].values[i]+1)] = 0
```

```python
PERIOD = 1.2 # days

with pm.Model() as model:

    # The mean flux of the time series
    mean = pm.Normal("mean", mu=0., sigma=10.)

    # A jitter term describing excess white noise
    log_jitter = pm.Normal("log_jitter", mu=np.log(np.mean(lc['e'].values/np.nanmedian(lc['f']))), sigma=2.0)

    # A term to describe the non-periodic variability
    sigma = pm.InverseGamma(
        "sigma", **pmx.estimate_inverse_gamma_parameters(1.0, 5.0)
    )
    rho = pm.InverseGamma(
        "rho", **pmx.estimate_inverse_gamma_parameters(0.5, 2.0)
    )

    # The parameters of the RotationTerm kernel
    sigma_rot = pm.InverseGamma(
        "sigma_rot", **pmx.estimate_inverse_gamma_parameters(1.0, 5.0)
    )
    log_period = pm.Normal("log_period", mu=np.log(PERIOD), sigma=.1)
    period = pm.Deterministic("period", tt.exp(log_period))
    log_Q0 = pm.HalfNormal("log_Q0", sigma=2.0)
    log_dQ = pm.Normal("log_dQ", mu=0.0, sigma=2.0)
    f = pm.Uniform("f", lower=0.1, upper=1.0)

    # Set up the Gaussian Process model
    kernel = terms.SHOTerm(sigma=sigma, rho=rho, Q=1 / 3.0)
    kernel += terms.RotationTerm(
        sigma=sigma_rot,
        period=period,
        Q0=tt.exp(log_Q0),
        dQ=tt.exp(log_dQ),
        f=f,
    )
    gp = GaussianProcess(
        kernel,
        t=lc['t'].values[np.where((fmask > 0))[0]],
        diag=lc['e'].values[np.where((fmask > 0))[0]]**2 + tt.exp(2 * log_jitter),
        mean=mean,
        quiet=False,
    )

    # Compute the Gaussian Process likelihood and add it into the
    # the PyMC3 model as a "potential"
    gp.marginal("gp", observed=lc['f'].values[np.where((fmask > 0))[0]]/np.nanmedian(lc['f'])-1)

    # Compute the mean model prediction for plotting purposes
    pm.Deterministic("pred", gp.predict(lc['f'].values[np.where((fmask > 0))[0]]/np.nanmedian(lc['f'])-1))

    pm.Deterministic("pred2", gp.predict(lc['f'].values[np.where((fmask > 0))[0]]/np.nanmedian(lc['f'])-1, 
                                     t=lc['t'].values))


    # Optimize to find the maximum a posteriori parameters
    map_soln = pmx.optimize()
    
```

```python
plt.plot(lc['t'], lc['f']/np.nanmedian(lc['f'])-1)
plt.plot(lc['t'].values[np.where((fmask > 0))[0]], map_soln["pred"],marker='*')
plt.plot(lc['t'].values, map_soln["pred2"], color='g',marker='.',alpha=0.5)

# plt.xlim(697,698)
plt.xlim(649,652)
```

```python
# OK, so now lets wrap that up in a loop and do GPs on everything
```

```python
plt.figure(figsize=(7,6))
bigtime = np.zeros(len(fbeye))
big_lfl1 = np.zeros(len(fbeye))
big_lfl2 = np.zeros(len(fbeye))
big_lfl3 = np.zeros(len(fbeye))
big_Nfl = np.zeros(len(fbeye))

CC = plt.cm.get_cmap('RdBu_r')

tbig = np.array([])
fbig = np.array([])
ebig = np.array([])
gbig = np.array([])

for k in range(len(fbeye)):
    ff = pd.read_table(fbeye[k], skiprows=11, names = cnames, delim_whitespace=True)
    lc = pd.read_csv(fbeye[k][:-6], names=['t','f','e'])

    TOTEXP = np.nanmedian(np.diff(lc['t'])) * len(lc)
    big_Nfl[k] = len(ff) / TOTEXP
    
    medflux = np.nanmedian(lc['f'])

    fmask = np.ones_like(lc['t'].values, dtype=int)
    for i in range(len(ff)):
        fmask[int(ff['Start_INDX'].values[i]):int(ff['Stop_INDX'].values[i]+1)] = 0


    PERIOD = 1.2 # days

    with pm.Model() as model:

        # The mean flux of the time series
        mean = pm.Normal("mean", mu=0., sigma=10.)

        # A jitter term describing excess white noise
        log_jitter = pm.Normal("log_jitter", mu=np.log(np.mean(lc['e'].values/medflux)), sigma=2.0)

        # A term to describe the non-periodic variability
        sigma = pm.InverseGamma(
            "sigma", **pmx.estimate_inverse_gamma_parameters(1.0, 5.0)
        )
        rho = pm.InverseGamma(
            "rho", **pmx.estimate_inverse_gamma_parameters(0.5, 2.0)
        )

        # The parameters of the RotationTerm kernel
        sigma_rot = pm.InverseGamma(
            "sigma_rot", **pmx.estimate_inverse_gamma_parameters(1.0, 5.0)
        )
        log_period = pm.Normal("log_period", mu=np.log(PERIOD), sigma=.1)
        period = pm.Deterministic("period", tt.exp(log_period))
        log_Q0 = pm.HalfNormal("log_Q0", sigma=2.0)
        log_dQ = pm.Normal("log_dQ", mu=0.0, sigma=2.0)
        f = pm.Uniform("f", lower=0.1, upper=1.0)

        # Set up the Gaussian Process model
        kernel = terms.SHOTerm(sigma=sigma*2, rho=20, tau=9) # non-periodic piece
        kernel += terms.RotationTerm(
            sigma=sigma_rot,
            period=period,
            Q0=tt.exp(log_Q0),
            dQ=tt.exp(log_dQ),
            f=f,
        )
        gp = GaussianProcess(
            kernel,
            t=lc['t'].values[np.where((fmask > 0))[0]],
            diag=lc['e'].values[np.where((fmask > 0))[0]]**2 + tt.exp(2 * log_jitter),
            mean=mean,
            quiet=True,
        )

        # Compute the Gaussian Process likelihood and add it into the
        # the PyMC3 model as a "potential"
        gp.marginal("gp", observed=lc['f'].values[np.where((fmask > 0))[0]]/np.nanmedian(lc['f'])-1)

        # Compute the mean model prediction for plotting purposes
        pm.Deterministic("pred", gp.predict(lc['f'].values[np.where((fmask > 0))[0]]/medflux-1))

        pm.Deterministic("pred2", gp.predict(lc['f'].values[np.where((fmask > 0))[0]]/medflux-1, 
                                         t=lc['t'].values))

        # Optimize to find the maximum a posteriori parameters
        map_soln = pmx.optimize()
        
    tbig = np.append(tbig, lc['t'].values)
    fbig = np.append(fbig, lc['f'].values/medflux-1)
    ebig = np.append(ebig, lc['e'].values)
    gbig = np.append(gbig, map_soln["pred2"])
    
    ED = np.zeros(len(ff), dtype=float)
    EDerr = np.zeros(len(ff), dtype=float)
    for l in range(len(ED)):
        ED[l], EDerr[l] = EquivDur(lc['t'].values, lc['f'].values/medflux-1,
                                   lc['e'].values/medflux, 
                                   lc['t'].values, map_soln["pred2"], 
                                   int(ff['Start_INDX'][l]),int(ff['Stop_INDX'][l]))
    ss = np.argsort(np.array(ED))[::-1]
    ffd_x, ffd_y, ffd_xerr, ffd_yerr = FFD(ED, TOTEXP=TOTEXP, 
                                       Lum=0., fluxerr=np.nanmedian(lc['e']), 
                                       dur = ff['t_stop'] - ff['t_start'])

    plt.plot(ffd_x, ffd_y, c=CC(np.nanmedian(lc['t'])/1600), lw=2)

    # write outputs!!!
    df = pd.DataFrame(data={'tstart':ff['t_start'][ss], 'tstop':ff['t_stop'][ss], 
                        'ED':ED[ss], 'EDerr':EDerr[ss],
                        'ffd_x':ffd_x, 'ffd_y':ffd_y, 
                        'ffd_xerr':ffd_xerr, 'ffd_yerr':ffd_yerr})

    df.to_csv(fbeye[k]+'_C.csv', index=False, index_label=False)


    bigtime[k] = np.nanmedian(lc['t'])
    big_lfl1[k] = np.log10(np.sum(ff['Equiv_Dur'].values) / (TOTEXP*86400))
    big_lfl2[k] = np.log10(np.sum(ED) / (TOTEXP*86400))
    big_lfl3[k] = np.log10(np.sum(np.trapz(lc['f'].values/medflux-1 - map_soln["pred2"], 
                                           lc['t'].values*60*60*24.) / (TOTEXP*86400)))
plt.xlabel('log Equivalent Duration (s)')
plt.ylabel('log Flare Rate (per day)')
plt.savefig('../figures/KIC_8507979_keplerFFDC.pdf', dpi=150, bbox_inches='tight', pad_inches=0.25, facecolor='w')
```
