```python
import matplotlib
matplotlib.rcParams.update({'font.size':18})
matplotlib.rcParams.update({'font.family':'serif'})
```

```python
# df = pd.read_csv(db)
tbl = Table.read(db, format='ascii.fixed_width', 
                names = ('Flare','Date','Start','Peak','End','Dur','Pk','Cts',
                         'Energy','XPos','YPos','Rad','AR','Flags'))
df = tbl.to_pandas()
```

```python
plt.scatter(df['Pk'], df['Cts'], s=1)
plt.yscale('log')
plt.xscale('log')
```

```python
plt.figure()
for k in range(100):
    x0 = k*1000
    x1 = (k+1)*1000
    flares = df['Cts'].values[x0:x1]
    ss = np.argsort(flares)[::-1]
    plt.plot(flares[ss], np.arange(len(flares)), 
             c='k', lw=0.5, alpha=0.5)
plt.xscale('log')
plt.yscale('log')
```

```python

df['iso'] = df['Date'].str.split('-', expand=True)[[2,1,0]].agg('-'.join, axis=1).astype('datetime64[ns]')
```

```python
plt.figure(figsize=(9,4))
plt.scatter(df['iso'], df['Cts'], s=1)
plt.ylim(0,1e7);
plt.ylabel('Total Counts')
```

```python
plt.figure(figsize=(9,4))
plt.scatter(df['iso'], df['Pk'], s=1)
plt.ylim(1e2,1e5);
plt.ylabel('Total Counts')
plt.yscale('log')
```

```python
# number of flares roughly per month
plt.figure(figsize=(9,4))
_ = plt.hist(df['iso'],bins=218)
```

```python
mjd = Time(df['iso']).mjd
tspan = 27.
nsteps = int((mjd[-1] - mjd[0]) / tspan)
# plt.cm.PiYG
CC = plt.cm.get_cmap('RdBu_r')

plt.figure(figsize=(8,7))
for k in range(100):
    fk = (mjd >= (mjd[0] + k*tspan)) & (mjd < (mjd[0] + (k+1)*tspan))
    ss = np.argsort(df['Cts'].values[fk])[::-1]
    totdur = max(mjd[fk]) - min(mjd[fk])
    plt.plot(df['Cts'].values[fk][ss], np.arange(len(ss))/totdur, 
         c=CC(float(k)/100), lw=1, alpha=1)

plt.yscale('log')
plt.xscale('log')
plt.xlabel('Total X-Ray Counts')
plt.ylabel('Cumulative Flares per Day')
# plt.title('RHESSI (2002-2009)')
plt.savefig('../figures/rhessi_FFD.pdf', dpi=150, bbox_inches='tight', pad_inches=0.25, facecolor='w')
```

```python
# borrowing from this example: https://stackoverflow.com/a/23062183

plt.figure(figsize=(9,4))
n, bins, patches = plt.hist(df['iso'],bins=218, color='k',alpha=0.25)
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# scale values to interval [0,1]
col = bin_centers - min(bin_centers)
col /= 100

for k in range(100):
    plt.setp(patches[k], 'facecolor', CC(k/100), alpha=1)

plt.ylabel('# Flares')
plt.savefig('../figures/rhessi_hist.pdf', dpi=150, bbox_inches='tight', pad_inches=0.25, facecolor='w')
```
